## Simple Linear Regression

```python
import math
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
x = df2.iloc[:,0]
x
```

```python
y = df2.iloc[:,1]
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
x_train = np.array(x_train).reshape(-1,1)
x_train
```

```python
x_test = np.array(x_test).reshape(-1,1)
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
c=lr.intercept_
m=lr.coef_
m
```

```python
c
```

```python
y_pred_train = lr.predict(x_train)
y_pred_train
```

```python
import matplotlib as lib
plt.scatter(x_train,y_train)
plt.plot(x_train,y_pred_train, color='red')
```

```python
y_pred_test = lr.predict(x_test)
y_pred_test
```

```python
var=np.array(4.5).reshape(-1,1)
y_pred_var = lr.predict(var)
y_pred_var
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np

threshold = 0.5
y_pred_binary = np.where(y_pred_test >= threshold, 1, 0)

cm = confusion_matrix(y_test, y_pred_binary)

import seaborn as sn
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 4))
sn.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

# In Lab - Keerthika Mam

```python
sales = pd.read_csv('/content/Annualsales.csv')
sales
```

```python
y = sales['sales']
x = sales.stores
```

```python
plt.figure(figsize=(20,10))

plt.plot(x,y,'o',markersize = 15)

plt.ylabel('Annual Net Sales', fontsize = 30)
plt.xlabel('Number of Stories in the District', fontsize = 30)
```

## Simple Linear Regression:

```python
m, b = np.polyfit(x,y,1)
print('The slope of line is {:.2f}.'.format(m))
print('The y-intercept is {:.2f}.'.format(b))
print('The best fit Simple Linear Regression line is {:.2f}x + {:.2f}.'.format(m,b))
```

```python
y_mean = y.mean()
x_mean = x.mean()

print('The centroid for this dataset is x = {:.2f} and y = {:.2f}'.format(x_mean, y_mean))
```

```python
plt.figure(figsize=(20,10))
plt.plot(x,y,'o',markersize = 14, label = 'Annual Net Sales')

plt.ylabel('Annual Net Sales', fontsize = 30)
plt.xlabel('Number of Stories in the District', fontsize = 30)

plt.plot(x_mean,y_mean,'*', markersize = 30, color = 'r')

plt.plot(x, m*x + b, '-', label = 'Simple Linear Regression Line', linewidth = 4)

plt.ylabel('Annual Net Sales', fontsize = 30)
plt.xlabel('Number of Stores in District', fontsize = 30)

plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.annotate('Centroid',
             xy=(x_mean-0.1, y_mean-5),
             xytext=(x_mean-3, y_mean-20),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize = 30)

plt.legend(loc = 'upper right', fontsize = 20)
```

## Prediction

```python
def predict(query):
  if query >= 1:
    predict = m* query + b
    return predict
  else:
    print("You must have at least 1 store in the district to predict the annual net sales")
```

```python
predict(13)
```
