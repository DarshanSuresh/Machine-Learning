> **Find S**



1.Read the training data from a csv file


```python
#importing necessary libraries
import pandas as pd

#importing the data
df = pd.read_csv('/content/drive/MyDrive/Machine Learning Lab/Dataset/set-1.csv')
df
```

2.Display the data frame information

```python
df.info()
```

3.Show unique values from all the attributes

```python
for i in df.columns:
    print(i,':',df[i].unique())
```

```python
target = df['Poisonous'].values
target
```

```python
concepts = df[df.columns[df.columns != 'Poisonous']].values
concepts
```

4.Setting initial hypothesis to most specific hypothesis

```python
#Setting initial hypothesis to most specific hypothesis
initial_hypothesis = ['0']*concepts.shape[1]
initial_hypothesis
```

5.Implementing of Find S algorithm

```python
#Find-S Algorithm
rows,columns = concepts.shape
def Find_S(initial_hypothesis):
    for i in range(rows):
        if target[i]=='Yes':
            for j in range(columns):
                if initial_hypothesis[j] == '0':
                    initial_hypothesis[j] = concepts[i][j]
                elif initial_hypothesis[j] != concepts[i][j]:
                    initial_hypothesis[j] = '?'
        print('Step', i+1, ':Hypothesis', initial_hypothesis)
    return initial_hypothesis
```

6.Displaying final hypothesis

```python
#Final_hypothesis
final_hypothesis = Find_S(initial_hypothesis)
print('Final Hypothesis is: ',final_hypothesis)
```

7.Displaying final hypothesis for  Poisonous and Non-Poisonous

```python
# Display final hypothesis for Poisonous and Non-Poisonous
poisonous_hypothesis = []
non_poisonous_hypothesis = []

for i in range(len(target)):
  if target[i] == 'Yes':
    poisonous_hypothesis.append(concepts[i])
  else:
    non_poisonous_hypothesis.append(concepts[i])

print("Final Hypothesis for Poisonous:")
if poisonous_hypothesis:
  for hypothesis in poisonous_hypothesis:
    print(hypothesis)
else:
  print("No Poisonous examples found.")

print("\nFinal Hypothesis for Non-Poisonous:")
if non_poisonous_hypothesis:
  for hypothesis in non_poisonous_hypothesis:
    print(hypothesis)
else:
  print("No Non-Poisonous examples found.")
```

# **SET-3**


> **Naive Bayes**



```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Step 1: Generate the dataset
n_features = 8
n_classes = 2
n_samples = 800
n_informative = 2
random_state = 1
n_clusters_per_class = 1

X, y = make_classification(n_samples=n_samples,
                           n_features=n_features,
                           n_informative=n_informative,
                           n_classes=n_classes,
                           n_clusters_per_class=n_clusters_per_class,
                           random_state=random_state)
```

```python
# Step 2: Plot the data (using only the informative features for visualization)
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title('Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Class')
plt.grid()
plt.show()
```

```python
# Step 3: Split the dataset into training and testing sets with 30% test size
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_state)
```

```python
# Step 4: Build the Naive Bayes classifier and train the model
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
# Step 5: Predict values for test samples
y_pred = model.predict(X_test)
```

```python
# Step 6: Estimate performance metrics
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

```python
# Step 7: Plot the confusion matrix
plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=model.classes_).plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

1.	Read dataset and visualize the mass and width features using scatterplot.

```python
#Load the dataset and read first 5 data samples
import pandas as pd
df= pd.read_csv("/content/drive/MyDrive/Machine Learning Lab/Dataset/fruits.csv", usecols = ['mass', 'width','height'])
df.head()
```

```python
#Visualize the Data
import seaborn as sns
sns.scatterplot(data = df, x = 'mass', y = 'width', hue = 'height')
```

2.	Setup the training and test splits.


```python
#set up training and test splits
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['mass', 'width']], df[['height']], test_size=0.33, random_state=0)
```

```python
#normalize the training and test data using the preprocessing.normalize() method
from sklearn import preprocessing
X_train_norm = preprocessing.normalize(X_train)
X_test_norm = preprocessing.normalize(X_test)
```

3.	Find the optimal number of clusters for given data for different k values.

```python
#Fitting and Evaluating the Model
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(X_train_norm)
```

```python
#visulaize the k=3 clusters
sns.scatterplot(data = X_train, x = 'mass', y = 'width', hue = kmeans.labels_)
```

4.	Plot the performance score to choose optimal k.

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_train_norm, kmeans.labels_, metric='euclidean')
```

5.	Build the model with optimal k.


```python
#Choosing the best number of clusters
K = range(2, 8)
fits = []
score = []
for k in K:
    # train the model for current value of k on training data
    model = KMeans(n_clusters = k, random_state = 0, n_init='auto').fit(X_train_norm)

    # append the model to fits
    fits.append(model)

    # Append the silhouette score to scores
    score.append(silhouette_score(X_train_norm, model.labels_, metric='euclidean'))
```

6.	Generate the visualization of k clusters

```python
#plot the performance score for different K values to apply elbow method
sns.lineplot(x = K, y = score)
```
